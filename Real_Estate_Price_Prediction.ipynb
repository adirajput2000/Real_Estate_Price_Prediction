```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df2 = df1.drop(['area_type','society','balcony','availability'], axis='columns')
```

```python
df3 = df2.dropna()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0])) 
```

```python
df3.head()
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    token = x.split('-')
    if len(token) == 2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2660-2700')
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
```

```python
df4['total_sqft'].unique()
```

```python
df5 = df4.copy()
```

```python
df5['price_per_sqft'] = df5['price'] * 100000 / df5['total_sqft']
df5.head()
```

```python
len(df5['location'].unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())
```

```python
df5.head()
```

```python
location_stat = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_stat
```

```python
len(location_stat[location_stat<=10])
```

```python
location_stat_less_than_10 = location_stat[location_stat<=10]
location_stat_less_than_10
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stat_less_than_10 else x)
```

```python
df5[df5.total_sqft/df5.bhk<300]
```

```python
df6 = df5[~(df5.total_sqft/df5.bhk<300)]
```

```python
df6
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out=pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
```

```python
df7 = remove_pps_outliers(df6)
df7
```

```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location == location) & (df.bhk==2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft, bhk2.price_per_sqft, color='blue', label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price_per_sqft, marker='+', color='green', label='3BHK',s =50)
    plt.xlabel('Total Square Feet Area')
    plt.ylabel('Price Per Square Feet Area')
    plt.title(location)
    plt.legend()
```
